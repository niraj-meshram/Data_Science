```python
import pandas as pd
import numpy as np
import seaborn as se

df = pd.read_csv("bhp.csv")
df
```

```python
lower_bound = df["price_per_sqft"].quantile(0.001)
upper_bound = df["price_per_sqft"].quantile(0.999)
lower_bound,upper_bound
```

```python
df_no_outiler_using_percentile = df[(df["price_per_sqft"] >= lower_bound) & (df["price_per_sqft"] <= upper_bound)]
df_no_outiler_using_percentile
```

```python
df
```

```python
df.describe()
```

```python
df_no_outiler_using_percentile.describe()
```

```python
mean = df_no_outiler_using_percentile.price_per_sqft.mean()
mean
```

```python
standard_deviation = df_no_outiler_using_percentile.price_per_sqft.std() 
standard_deviation
```

```python
mean - 4*standard_deviation
```

```python
mean + 4*standard_deviation
```

```python
df_no_outiler_using_standard_deviation = df_no_outiler_using_percentile[(df_no_outiler_using_percentile["price_per_sqft"] > -9900.429065502582) & (df_no_outiler_using_percentile["price_per_sqft"] < 23227.73653589432)]
df_no_outiler_using_standard_deviation
```

```python
df_no_outiler_using_standard_deviation.describe()
```

```python
se.histplot(df_no_outiler_using_standard_deviation.price_per_sqft,kde=True)
```

```python
df_no_outiler_using_standard_deviation["z_score"] = ((df_no_outiler_using_standard_deviation.price_per_sqft-df_no_outiler_using_standard_deviation.price_per_sqft.mean())/df_no_outiler_using_standard_deviation.price_per_sqft.std())
df_no_outiler_using_standard_deviation
```

```python
df_no_outiler_using_standard_deviation = df_no_outiler_using_standard_deviation[(df_no_outiler_using_standard_deviation["z_score"] > -4) & (df_no_outiler_using_standard_deviation["z_score"] < 4)]
df_no_outiler_using_standard_deviation
```
