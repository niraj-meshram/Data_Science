```python
import pandas as pd
import numpy as np
import seaborn as se

df = pd.read_csv("logincome.csv",index_col=None,names=["income","count"],skiprows=1)
df.head()
```

```python
se.set(rc={'figure.figsize':(11.7,8.27)})
g = se.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), 
                  rotation=45, 
                  horizontalalignment='right');
```

```python
se.set(rc={'figure.figsize':(11.7,8.27)})
g = se.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale="log");
```
