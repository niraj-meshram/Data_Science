```python
import pandas as pd
import numpy as np
df = pd.read_csv("movie_revenues.csv")
df.head()

df["revenue_in_million"] = df["revenue"].apply(lambda x: x/1000000)
df.revenue_in_million.describe()
```

```python
_, mean, std,*_= df.revenue_in_million.describe()
```

```python
mean
```

```python
std
```

```python
def get_z_score(revenue, mean, std):
    return (revenue-mean)/std
```

```python
df["z_score"] = df.revenue_in_million.apply( lambda x: get_z_score(x, mean, std))
df.head(3)
```

```python
df[df.z_score>3]
```

```python
def get_MAD(revenue_in_million):
    median = np.median(revenue_in_million)
    diff = abs(revenue_in_million-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_MAD(df.revenue_in_million)
median = np.median(df.revenue_in_million)
MAD,median
```

```python
def get_MAD_Z_Score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_MAD_Z_Score(2711, median, MAD)
```

```python
df["mad_z_score"] = df.revenue_in_million.apply( lambda x: get_MAD_Z_Score(x,median, MAD))
df.head(3)
```

```python
df[df.mad_z_score > 3]
```
