```python
import numpy as np
```

```python
y_predict = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
y_predict
```

```python
# Mean Absolute Error
```

```python
def mae(y_true,y_predict):
    total_error = 0
    for yt, yp in zip(y_true,y_predict):
        print(yt, yp)
        total_error += abs(yt-yp)
    mae = total_error/len(y_true)
    print("MAE:" ,mae)
    print("Total error:" ,total_error)
```

```python
mae(y_true,y_predict)
```

```python
mae = np.mean(np.abs(y_true-y_predict))
mae           
```

```python
# Mean Squared Error
```

```python
def mse(y_true,y_predict):
    total_error = 0
    for yt, yp in zip(y_true,y_predict):
        print(yt, yp)
        total_error += np.square(yt-yp)
    mse = total_error/len(y_true)
    print("MSE:" ,mse)
    print("Total error:" ,total_error)
```

```python
mse(y_true,y_predict)
```

```python
mse = np.mean(np.square(y_true-y_predict))
mse
```

```python
def log_loss(y_true,y_predict):
    total_error = 0
    y_predict_new = [max(i, eplison) for i in y_predict]
    y_predict_new = [min(i, 1-eplison) for i in y_predict_new]
    y_predict_new = np.array(y_predict_new)
    total_error = -np.mean(y_true*np.log(y_predict_new)+(1-y_true)*np.log(1-y_predict_new))
    return total_error
```
