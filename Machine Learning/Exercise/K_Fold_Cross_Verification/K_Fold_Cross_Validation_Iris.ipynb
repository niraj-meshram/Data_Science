```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
y_train_scaled = scalar.fit_transform(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
cross_val_score(LogisticRegression(max_iter=1000),iris.data,iris.target, cv=5,scoring="accuracy")
```

```python
from numpy import mean
mean(cross_val_score(LogisticRegression(max_iter=1000),iris.data,iris.target, cv=5,scoring="accuracy"))
```

```python
# from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
cross_val_score(SVC(kernel="linear", C = 1.0),iris.data,iris.target, cv=5,scoring="accuracy")
```

```python
mean(cross_val_score(SVC(kernel="linear", C = 1.0),iris.data,iris.target, cv=5,scoring="accuracy"))
```

```python
from sklearn.ensemble import RandomForestClassifier
cross_val_score(RandomForestClassifier(),iris.data,iris.target,cv=5, scoring="accuracy")
```

```python
mean(cross_val_score(RandomForestClassifier(),iris.data,iris.target,cv=5, scoring="accuracy"))
```
