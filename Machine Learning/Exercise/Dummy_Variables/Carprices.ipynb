```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv("carprices.csv") 
df
```

```python
%matplotlib inline
plt.title('Mileage vs Sell Price by Car Model')
plt.xlabel('Mileage')
plt.ylabel('Sell Price ($)')
sns.scatterplot(x='Mileage', y='Sell Price($)', hue='Car Model', data=df)
plt.show()
```

```python
%matplotlib inline
plt.title("Mileage vs Age(yrs)")
plt.xlabel("Mileage")
plt.ylabel("Age(yrs)")
sns.scatterplot(x="Mileage", y="Age(yrs)", hue="Car Model",data=df)
plt.show()
```

```python
%matplotlib inline
plt.title("Sell Price($) vs Age(yrs)")
plt.xlabel("Sell Price($)")
plt.ylabel("Age(yrs)")
sns.scatterplot(x="Sell Price($)",y="Age(yrs)", hue="Car Model", data=df)
plt.show()
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
merged = pd.concat([df,dummies], axis = "columns")
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"], axis = "columns")
final
```

```python
X = final.drop("Sell Price($)", axis = "columns")
X
```

```python
final
```

```python
X
```

```python
Y = final["Sell Price($)"]
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
feature_names = ["Mileage",	"Age(yrs)",	"Audi A5",	"BMW X5"]
```

```python
model.predict(pd.DataFrame([[90000, 25, True, False]],columns = feature_names))
```

```python
model.score(X,Y)
```
