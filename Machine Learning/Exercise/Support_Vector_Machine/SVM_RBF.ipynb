```python
data = datasets.load_digits()
```

```python
data.target_names
```

```python
X = data.data
```

```python
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_trained_scaled = scaler.fit_transform(X_train)
X_test_scaled =  scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
svm_rbf =SVC(kernel="rbf", C = 1.0, gamma = "scale")
svm_rbf.fit(X_trained_scaled,y_train)
y_pred_rbf = svm_rbf.predict(X_test_scaled)
accuracy_rbf = accuracy_score(y_test,y_pred_rbf)
print(f"RBF Kernel Accuracy: {accuracy_rbf:.2f}")
```

```python
svm_linear =  SVC(kernel="linear", C = 1.0)
svm_linear.fit(X_trained_scaled,y_train)
y_pred_linear =  svm_linear.predict(X_test_scaled)
accuracy_linear = accuracy_score(y_test,y_pred_linear)
print(f"Linear Kernel Accuracy: {accuracy_linear:.2f}")
```

```python
def make_it_fancy(func):
    def wrapper():
        print("✨✨")
        func()
        print("✨✨")
    return wrapper

@make_it_fancy
def say_hello():
    print("Hello!")

say_hello()
```

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"⏱️ Took {end - start:.2f} seconds")
    return wrapper

@timer
def slow_function():
    time.sleep(2)
    print("Done sleeping!")

slow_function()
```
