```python
import pandas as pd
import numpy as np

dataset =  pd.read_csv("Melbourne_housing_FULL.csv")
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
columns_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[columns_to_use]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
columns_to_fill_zero = ["Propertycount", "Distance", "Bedroom2", "Bathroom", "Car"]
dataset[columns_to_fill_zero] = dataset[columns_to_fill_zero].fillna(0)
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset["Landsize"] = dataset["Landsize"].fillna(dataset["Landsize"].mean)
dataset["BuildingArea"] = dataset["BuildingArea"].fillna(dataset["BuildingArea"].mean)
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first =True)
```

```python
dataset.head()
```

```python
X = dataset.drop("Price", axis = 1)
y = dataset["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol = 1.0)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn import linear_model
ridge_reg = linear_model.Ridge(alpha=50, max_iter=100, tol = 1.0)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
