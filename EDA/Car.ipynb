```python
#Importing the required libraries for EDA
import pandas as pd
import numpy as nm
import seaborn as se #visualisation
import matplotlib.pyplot as pypl#visualisation
```

```python
#Loading the data into the data frame
df = pd.read_csv("car_data.csv")
df.head(5)
```

```python
#Checking the types of data
df.dtypes
```

```python
#Describing the data
df.describe()
```

```python
#Dropping irrelevant columns
df=df.drop(["Engine Fuel Type","Number of Doors"], axis=1)
df
```

```python
#Renaming the columns
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders"})
df
```

```python
df.shape
```

```python
#Dropping the duplicate rows
duplicate_rows_df = df[df.duplicated()]
duplicate_rows_df
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(25)
```

```python
df.shape
```

```python
df.count()
```

```python
#Dropping the missing or null values.
df.isnull().sum()
```

```python
df =  df.dropna()
df.count()
```

```python
#Detecting Outliers
se.boxplot(x=df["MSRP"])
```

```python
se.boxplot(x=df["HP"])
```

```python
se.boxplot(x=df['Cylinders'])
```

```python
# This will calculate the quantile only for numeric columns
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
IQR
```

```python
df.dtypes
df.shape
```

```python
# Step 1: Identify numeric columns
# numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
# outliers_lower = numeric_df < lower_bound
# outliers_upper = numeric_df > upper_bound
# outliers = outliers_lower | outliers_upper

# df_cleaned_numeric = numeric_df[~outliers.any(axis=1)]
# df_cleaned_numeric

df = df[~((df.select_dtypes(include=['number']) < (Q1 - 1.5 * IQR)) | (df.select_dtypes(include=['number']) > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

```python
#Plot different features against one another (scatter), against frequency (histogram)

df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
pypl.title("Number of cars by make")
pypl.ylabel('Number of cars')
pypl.xlabel('Make')
```

```python
#Heat Maps
pypl.figure(figsize=(10,8))
c= df.select_dtypes(include=['number']).corr()
se.heatmap(c,cmap="BrBG",annot=True)
c
```
